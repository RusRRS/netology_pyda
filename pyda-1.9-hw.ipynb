##### Задание 1
Напишите функцию date_range, которая возвращает список дней между датами start_date и end_date.  
Даты должны вводиться в формате YYYY-MM-DD.

```python
from datetime import datetime # импорт класса из библиотеки 
from datetime import timedelta
```

```python
def date_range(start_date, end_date):
    # Переводим в нужный формат
    start_date_datetime = datetime.strptime(start_date, '%Y-%m-%d')
    end_date_datetime = datetime.strptime(end_date, '%Y-%m-%d')    

    current_datetime = start_date_datetime
    current_datelist = []
    
    while current_datetime <= end_date_datetime:
        current_datelist.append(current_datetime.strftime('%Y-%m-%d'))
        current_datetime += timedelta(days=1)
    
    return current_datelist
    
date_range('2018-01-02', '2018-01-07')
```

В последнем примере поиска по словарю мы использовали 3 столбца.  
Напишите функцию, которая формирует словарь для поиска по **n** столбцам.

```python
def stats_dict_func(filepath, n):
    stats_dict_n = {} # Задать пустой список
    with open(filepath) as f:
        for line in f:
            line = line.strip().split(',')
            stats_dict_n[tuple(line[:n-1])] = line[n-1:]
    return stats_dict_n

stats_dict_func('F:\Program\git_base\GitHub\MyRepo\PYDA_09\stats.csv', 4)
```
