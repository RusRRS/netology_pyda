**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. 

Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

### Выполнение расчета выполняется исходя из следующих допущений:
| # | Знак зодиака | Дата начала   | Дата начала |
|-- |--------------|---------------|-------------|
|♈| Овен	      |21 марта |	 20 апреля|
|♉|Телец	   |21 апреля 	 |21 мая|
|♊|Близнецы	  |22 мая 	| 21 июня|
|♋|Рак	    |22 июня 	| 22 июля|
|♌|Лев	    |23 июля 	| 21 августа|
|♍|Дева	    |22 августа |	 23 сентября|
|♎|Весы	    |24 сентября |	 23 октября|
|♏|Скорпион	|24 октября |	 22 ноября|
|♐|Стрелец	|23 ноября 	| 22 декабря|
|♑|Козерог	|23 декабря |	 20 января|
|♒|Водолей	|21 января 	| 19 февраля|
|♓|Рыбы	    |20 февраля 	| 20 марта|

```python
# для начала получим месяц:
birth_month_input = input("Введите месяц:") # переменная для получения значения месяца
birth_month = birth_month_input.upper() # оставляем возможность написания с большой буквы

# Теперь получим дату:
try:
    birth_date = int(input("Введите число:"))
    # Убираем несуществующие даты:
    # Дата больше 31
    if birth_date > 31:
        print("Ошибка ввода данных")
    # Дата 29 и 30 февраля
    elif birth_month == "ФЕВРАЛЬ" and birth_date > 28:
        print("Ошибка ввода данных")
    # Дата 30 апреля, июня, сентября и ноября
    elif birth_date == 31 and (birth_month == "АПРЕЛЬ" or \
                               birth_month == "ИЮНЬ" or \
                               birth_month == "СЕНТЯБРЬ" or \
                               birth_month == "НОЯБРЬ"):
        print("Ошибка ввода данных")
    
    # После того, как убрали несуществующие даты, определяем знаки зодиака:        
    
    elif birth_month == "ЯНВАРЬ":
        # Козерог или Водолей    
        if birth_date < 21:
            print("Козерог")
        else:
            print("Водолей")

    elif birth_month == "ФЕВРАЛЬ":
        # Водолей или Рыбы    
        if birth_date <20:
            print("Водолей")
        else:
            print("Рыбы") 

    elif birth_month == "МАРТ":
        # Рыбы или Овен
        if birth_date < 21:
            print("Рыбы")
        else:
            print("Овен") 

    elif birth_month == "АПРЕЛЬ":
        # Овен или Телец
        if birth_date < 21:
            print("Овен")
        else:
            print("Телец")

    elif birth_month == "МАЙ":
        # Телец или Близнецы    
        if birth_date <22:
            print("Телец")
        else:
            print("Близнецы")

    elif birth_month == "ИЮНЬ":
        # Близнецы или Рак    
        if birth_date <22:
            print("Близнецы")
        else:
            print("Рак")       

    elif birth_month == "ИЮЛЬ":
        # Рак или Лев    
        if birth_date < 23:
            print("Рак")
        else:
            print("Лев")         

    elif birth_month == "АВГУСТ":
        # Лев или Дева    
        if birth_date < 22:
            print("Лев")
        else:
            print("Дева")          

    elif birth_month == "СЕНТЯБРЬ":
        # Дева или Весы    
        if birth_date <24:
            print("Дева")
        else:
            print("Весы")             

    elif birth_month == "ОКТЯБРЬ":
        # Весы или Скорпион    
        if birth_date < 24:
            print("Весы")
        else:
            print("Скорпион")          

    elif birth_month == "НОЯБРЬ":
        # Скорпион или Стрелец
        if birth_date < 23:
            print("Скорпион")
        else:
            print("Стрелец")        

    elif birth_month == "ДЕКАБРЬ":
        # Стрелец или Козерог    
        if birth_date < 23:
            print("Стрелец")
        else:
            print("Козерог")
    else:
        print("Ошибка ввода данных")        


except (TypeError, ValueError):
    print('Ошибка ввода данных')
```
